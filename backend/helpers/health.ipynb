```python
def extract_coordinates(point_str):
    # Remove the "POINT (" prefix and ")" suffix
    coords_str = point_str.replace("POINT (", "").replace(")", "")

    # Split the string by whitespace to separate the x and y coordinates
    x_str, y_str = coords_str.split()

    # Convert the coordinates from strings to floats
    x = float(x_str)
    y = float(y_str)

    return round(x, 3), round(y, 3)

# Apply the function to each row of the DataFrame and create a new column with tuples
df[['x', 'y']] = df['Geolocation'].apply(lambda x: pd.Series(extract_coordinates(x)))

# Combine x and y coordinates into a string in the format "x,y"
# df['coords'] = df.apply(lambda row: '{},{}'.format(row['x'], row['y']), axis=1)

df.head()
```
